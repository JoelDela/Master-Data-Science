# House Pricing

## Problem Statement

We would like to predict the house of a house based on its characteristics.

```python
%pylab inline
import pandas as pd
```

```python
df = pd.read_csv("house_prices.csv")
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
df=df[df["TotalSF"]<6000]
```

```python
# Data Preparation
X = df[['TotalSF']] # pandas DataFrame
y = df["SalePrice"] # pandas Series
```

```python
plt.scatter(X,y)
```

## Linear regression

```python
# Load the library
from sklearn.linear_model import LinearRegression
```

```python
# Create an instance of the model 
reg = LinearRegression()
```

```python
# Fit the regressor 
reg.fit(X,y)
```

```python
# Do predictions 
reg.predict([[2540],[3500],[4000]])
```

```python
# Load the library
from sklearn.model_selection import train_test_split
```

```python
# Create 2 groups each with input and labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10)
```

```python
# Fit only with training data
reg.fit(X_train,y_train)
```

```python
# Load the scorer
from sklearn.metrics import mean_absolute_error
# Use against predictions
mean_absolute_error(reg.predict(X_test),y_test)
```

```python
# Podemos calcular el MAE a mano
y_hat=reg.predict(X_test)
sum(np.abs(y_hat-y_test)/len(y_hat))
```

```python
# Podemos calcular el MAPE a mano
y_hat=reg.predict(X_test)
sum(np.abs(y_hat-y_test)/(len(y_hat)*y_test))
```

```python
# Load the scorer
from sklearn.metrics import mean_squared_error
# Use against predictions (we must calculate the square root of the MSE)
np.sqrt(mean_squared_error(reg.predict(X_test),y_test))
```

```python
# Podemos ver la ordenada en el origen
reg.intercept_
```

```python
# Podemos ver la pendiente
reg.coef_
```

```python
npoints=600
rango=6000

plt.scatter(X_train,y_train,color='b')
plt.scatter(X_test,y_test,c='g')
xkk=np.linspace(0,rango,npoints)
plt.plot(xkk,reg.coef_*xkk+reg.intercept_,c='r')
```

```python
# Se observa que si se transforman los datos se pueden obtener mejores resultados
plt.scatter(X_train**2,y_train,color='b')
```

## K-neighbors Regressor

```python
# Load the library 
from sklearn.neighbors import KNeighborsRegressor 
# Create an instance 
regk = KNeighborsRegressor(n_neighbors=2) 
# Fit the data 
regk.fit(X,y)
```

```python
# Load the scorer
from sklearn.metrics import mean_squared_error
# Use against predictions (we must calculate the square root of the MSE)
np.sqrt(mean_squared_error(regk.predict(X_test),y_test))
```

```python
xgrid = [ [i] for i in np.linspace(1000,5000,1000)]
plt.plot(xgrid,reg.predict(xgrid),color='red')
plt.plot(xgrid,regk.predict(xgrid),color='yellow')
plt.scatter(X,y)
```

## Cross Validation

```python
# Load the library
from sklearn.model_selection import cross_val_score
# We calculate the metric for several subsets (determine by cv)
# With cv=5, we will have 5 results from 5 training/test
cross_val_score(reg,X,y,cv=5,scoring="neg_mean_squared_error")
```

```python
np.sqrt(-np.mean(cross_val_score(reg,X,y,cv=10,scoring="neg_mean_squared_error")))
```

## GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor

reg_test = GridSearchCV(KNeighborsRegressor(),
                       param_grid={"n_neighbors":np.arange(3,100)},scoring='neg_mean_absolute_error')
# Fit will test all of the combinations
reg_test.fit(X,y)
```

```python
reg_test.best_score_
```

```python
reg_test.best_estimator_.predict([[1200]])
```

```python
reg_test.best_params_
```

## Decision Tree

```python
# Load the library
from sklearn.tree import DecisionTreeRegressor
# Create an instance
regd = DecisionTreeRegressor(max_depth=3)
# Fit the data
regd.fit(X,y)
```

## Correlacin

```python
# Direct Calculation 
np.corrcoef(regd.predict(X_test),y_test)
```

```python
# Custom Scorer 
from sklearn.metrics import make_scorer 
def corr(pred,y_test): 
    return np.corrcoef(pred,y_test)[0][1] 
# Put the scorer in cross_val_score. EL make_scorer espera que le des 2 vectores y que devuelva un valor 
cross_val_score(regd,X,y,cv=5,scoring=make_scorer(corr))
```

## Bias

```python
# Direct Calculation 
np.mean(reg.predict(X_test)-y_test)
# Custom Scorer 
from sklearn.metrics import make_scorer 
def bias(pred,y_test): 
    return np.mean(pred-y_test) 
# Put the scorer in cross_val_score 
cross_val_score(regd,X,y,cv=5,scoring=make_scorer(bias))
```
