# Basic Car Speed 

## Problem Statement

We would like to determine the speed of a car based on the bumpiness and slope of the road. For this, we have a list of the speed from a list of drivers.

```python
%pylab inline
import pandas as pd
```

```python
df = pd.read_csv("terrain.csv")
```

```python
X = df[["bumpiness","grade"]].values
y = df["target"]
```

```python
def draw(clf,X,y,h=0.05):
    
    plt.figure(figsize=(10,10))

    x_min, x_max = X[:, 0].min() - 0.05, X[:, 0].max() + .05
    y_min, y_max = X[:, 1].min() - .05, X[:, 1].max() + .05
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    if clf is not None:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.scatter(xx,yy,c=Z)
    else:
        plt.scatter(xx,yy)
    
    plt.scatter(X[:,0],X[:,1],c=y,cmap="Paired")
```

## Logistic Regression

```python
# Load the library
from sklearn.linear_model import LogisticRegression
# Create an instance of the classifier
clf=LogisticRegression()
# Fit the data
clf.fit(X,y)
```

```python
plt.scatter(df['bumpiness'],df['grade'],c=df['target'])
```

```python
# Load the library
from sklearn.model_selection import train_test_split
# Create 2 groups each with input and labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10)
```

```python
# Load the library
from sklearn.model_selection import cross_val_score
# We calculate the metric for several subsets (determine by cv)
# With cv=5, we will have 5 results from 5 training/test
cross_val_score(clf,X,y,cv=5,scoring="accuracy")
```

```python
# With Metrics
from sklearn.metrics import accuracy_score
accuracy_score(y_test,clf.predict(X_test))
```

```python
draw(clf,X,y,0.01)
```

```python
draw(clf,X_train,y_train,0.01)
```

```python
draw(clf,X_test,y_test,0.01)
```

## K-neighbors Classification

```python
# Load the library 
from sklearn.neighbors import KNeighborsClassifier 
# Create an instance 
regk = KNeighborsClassifier(n_neighbors=5)
# Fit the data 
regk.fit(X,y)
```

```python
from sklearn.model_selection import GridSearchCV

reg_test = GridSearchCV(KNeighborsClassifier(), cv=5,
                       param_grid={"n_neighbors":np.arange(3,50)},scoring='accuracy')
# Fit will test all of the combinations
reg_test.fit(X,y)
```

```python
reg_test.best_estimator_
```

```python
reg_test.best_score_
```

```python
draw(reg_test.best_estimator_,X,y,0.01)
```

```python
# Metrics
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import confusion_matrix, classification_report
precision_score(y_test,clf.predict(X_test))
```

```python
print(classification_report(y_test,clf.predict(X_test)))
```

```python
print(confusion_matrix(y_test,clf.predict(X_test)))
```

```python
# Cross Validation
cross_val_score(clf,X,y,scoring="precision")
```

```python
cross_val_score(clf,X,y,scoring="recall")
```

## SVM

```python
# Load the library
from sklearn.svm import SVC
# Create an instance of the classifier
clf = SVC(kernel="linear",C=10)
# Fit the data
clf.fit(X,y)
```

```python
cross_val_score(clf,X,y,scoring="recall").mean()
```

```python
from sklearn.model_selection import GridSearchCV
clfSVC=GridSearchCV(SVC(kernel='poly'),param_grid={'C':[1,10,100,1000,100000],'degree':[2,3,4,5]},scoring='recall')
clfSCV.fit(X,y)
```

```python
cross_val_score(clfSCV,X,y,scoring="recall").mean()
```

```python
draw(clfSCV.best_estimator_,X,y,0.01)
```

## Decision Tree Classification

```python
# Import library
from sklearn.tree import DecisionTreeClassifier
# Create instance
clf = DecisionTreeClassifier(min_samples_leaf=20,max_depth=3)
# Fit the data
clf.fit(X,y)
```

```python
draw(clf,X,y,0.01)
```

```python
from sklearn.model_selection import GridSearchCV
clf_test=GridSearchCV(DecisionTreeClassifier(),param_grid={'min_samples_leaf':np.arange(20,50),'max_depth':np.arange(1,20)},scoring='recall')
clf_test.fit(X,y)
```

```python
clf_test.best_estimator_
```

```python
draw(clf_test.best_estimator_,X,y,0.01)
```

```python
pred=clf_test.best_estimator_.predict_proba(X_test)
```

```python
print(pred)
```
