```python
import pandas as pd
```

```python
# Example taken from here:
# http://faculty.cas.usf.edu/mbrannick/regression/Reg2IV.html
df = pd.read_csv("../data/machine_performance.csv")
```

```python
df.head()
```

```python
X=df[["X1","X2"]].values
y=df["Y"].values
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth=5)
```

```python
model.fit(X,y)
```

```python
df["pred1"]=model.predict(X)
df["resid"]=df["Y"]-df["pred1"]
```

```python
df
```

```python
model2=DecisionTreeRegressor(max_depth=5)
```

```python
model2.fit(df[["X1","X2"]].values,df["resid"])
```

```python
df["pred2"]=model2.predict(df[["X1","X2"]])
df["resid2"]=df["Y"]-df["pred1"]-df["pred2"]
```

```python
df
```

```python
model3=DecisionTreeRegressor(max_depth=4)
model3.fit(df[["X1","X2"]].values,df["resid2"])
df["pred3"]=model3.predict(df[["X1","X2"]])
df["resid3"]=df["Y"]-df["pred1"]-df["pred2"]-df["pred3"]
```

```python
df
```

```python
df["final_pred"]=df["pred1"]+df["pred2"]+df["pred3"]
```

```python
df.head()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(df["Y"],df["pred1"])
```

```python
mean_squared_error(df["Y"],df["final_pred"])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lmodel = LinearRegression()
lmodel.fit(X,y)
```

```python
mean_squared_error(df["Y"],lmodel.predict(df[["X1","X2"]]))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rmodel=RandomForestRegressor()
rmodel.fit(df[["X1","X2"]].values,df["Y"])
```

```python
mean_squared_error(df["Y"],rmodel.predict(df[["X1","X2"]]))
```
