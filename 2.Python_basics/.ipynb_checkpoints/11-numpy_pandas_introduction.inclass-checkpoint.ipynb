### Function application and mapping

Function application and mapping allows us to modify the elements of a DataFrame (columns with apply or elements with applymap) without for loops. This way we are not constrained to the functions already implemented by pandas or numpy.

````python
df3*3
````

````python
np.sqrt(df3[['population','year']]) # Solo se puede aplicar sobre columnas numricas
````

Si queremos usar funciones propias en dataframes habr que utilizar los siguientes cdigos:

````python
def minimax(col):
    return col.max() - col.min()
````

__La funcin .apply aplicar la funcin sobre la columna. Por norma general, cualquier funcin que acte sobre una serie podr actuar sobre un dataframe, por lo que no es tan til__

````python
df3[['year','population']].apply(minimax)
````

````python
minimax(df3[['population','year']])
````

````python
df3.max()
````

__La funcin .applymap() es mucho ms til. Permite aplicar funciones arbitrarias elemento a elemento__

````python
df3.applymap(str).dtypes #Se observa que lo hace elemento a elemento.
````

This is a typical use case for lambdas (anonymous functions)

````python
df3.applymap(lambda cell: str(cell)[:2])
````

````python
df3[['population','year','debt']].applymap(lambda cell: '%.2f'%cell)
````

### Filtering out missing data

````python
x.dropna()
````

````python
nrows=10
w=pd.Series(np.random.choice([1,2,95,17,np.nan], size=nrows))
x=pd.Series(np.random.choice([1,2,3,5,7,np.nan], size=nrows))
y=pd.Series(np.random.choice([8,11,22,7,np.nan], size=nrows))
z=pd.Series(np.random.choice([6,10,3,12,7,np.nan], size=nrows))

df=pd.DataFrame([w,x,y,z],index=['w','x','y','z']).T
df
````

````python
df.dropna()
````

any() and all() are functions of boolean Series. They reduce the Series to a single boolean value by applying repeatedly the operators "or" and "and", respectively.

````python
df.dropna(subset=['w','z'],how='all') #Con el comando 'how' indicas como quieres descartar los NaN de las columnas seleccionadas
````

The thresh argument specifies the minimum number of non-null values required to keep a column (or row, with axis=1)

````python
df.dropna(thresh=3) #Estas indicando que exiges que al menos en la columna haya MNIMO 3+1 registros no nulos
````
