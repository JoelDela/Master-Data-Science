# Time Shifting

- Desplazar valores +-N filas

```python
import pandas as pd
```

```python
df = pd.read_csv('../UPDATE_TSA_NOTEBOOKS/Data/starbucks.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
# Desplaza los valores N veces hacia abajo
df.shift(2).head()
```

```python
# Desplaza los valores N veces hacia arriba
df.shift(-2).head()
```

```python
df.shift(-2).tail()
```
