# Visualizing TimeSeries

```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('../UPDATE_TSA_NOTEBOOKS/Data/starbucks.csv',index_col='Date',parse_dates=True)
df.head()
```

```python
df.index
```

### Ploteando Columnas

```python
# Ploteamos las dos columnas pero estn en diferentes escalas
df.plot(grid=True,legend=True,figsize=(12,6));
```

```python
# Close por separado
ax = df['Close'].plot(color='b',grid=True,legend=True,title='Closing Price',figsize=(12,6))
ax.autoscale(axis='both',tight=True)
ax.set(xlabel='Date',ylabel='Price');
```

```python
# volume por separado
ax= df['Volume'].plot(color='orange',grid=True,legend=True,title='Volume',figsize=(12,6))
ax.autoscale(axis='both',tight=True)
ax.set(xlabel='Date',ylabel='Volume Quantity');
```

# Xlim Ylim: Ploteando Periodos Concretos

- X limits

#### 1 Forma: Indexando

```python
# Ploteando solo los Closing Prices de 2017
ax = df['Close']['2017-01-01':'2017-12-31'].plot(color='b',grid=True,legend=True,title='Closing Price 2017',figsize=(12,6))
ax.autoscale(axis='both',tight=True)
ax.set(xlabel='Date',ylabel='Price');
```

#### 2 Forma: xlimit

```python
# Close por separado
ax = df['Close'].plot(color='b',grid=True,legend=True,title='Closing Price',figsize=(12,6))
ax.set(xlabel='Date',ylabel='Price')
ax.set(xlim=['2017-01-01','2017-12-31'],ylim=[50,65]);
```

```python
# Close por separado
ax = df['Close'].plot(color='b',grid=True,legend=True,title='Closing Price',figsize=(12,6),xlim=['2017-01-01','2017-12-31'],ylim=[50,65],ls='--', c='green')
ax.set(xlabel='Date',ylabel='Price');
```

# X Ticks

```python
from matplotlib import dates
```

```python
# Mirar los Xticks en el cambio de mes!!!
df['Close'].plot(xlim=['2017-01-01','2017-03-01'],ylim=[50,60],grid=True,figsize=(12,6));
```

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-03-01'],ylim=[50,60],grid=True,figsize=(12,6))
ax.set(xlabel='')
# SET THE TICK LOCATOR AND FORMATTER FOR THE MAJOR AXIS
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0));
```

Notice that dates are spaced one week apart. The dates themselves correspond with <tt>byweekday=0</tt>, or Mondays.<br>
For a full list of locator options available from <tt>matplotlib.dates</tt> visit <a href='https://matplotlib.org/api/dates_api.html#date-tickers'>https://matplotlib.org/api/dates_api.html#date-tickers</a>

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-03-01'],ylim=[50,60],grid=True,figsize=(12,6))
ax.set(xlabel='')
# SET THE TICK LOCATOR AND FORMATTER FOR THE MAJOR AXIS
ax.xaxis.set_major_locator(dates.DayLocator(bymonthday=[1,15], interval=1, tz=None));
```

### Date Formatting
Formatting follows the Python datetime <strong><a href='http://strftime.org/'>strftime</a></strong> codes.<br>
The following examples are based on <tt>datetime.datetime(2001, 2, 3, 16, 5, 6)</tt>:
<br><br>

<tr><td>%m</td><td>Month as a zero-padded decimal number.</td><td>02</td></tr>
<tr><td>%B</td><td>Month as locales full name.</td><td>February</td></tr>

#### MAJOR AXIS

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-12-31'],ylim=[50,65],grid=True,figsize=(12,6))
ax.set(xlabel='')
# SET THE TICK LOCATOR AND FORMATTER FOR THE MAJOR AXIS
ax.xaxis.set_major_locator(dates.MonthLocator(bymonth=None, bymonthday=1, interval=1, tz=None));
```

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-12-31'],ylim=[50,65],grid=True,figsize=(12,6))
ax.set(xlabel='')
# SET THE TICK LOCATOR AND FORMATTER FOR THE MAJOR AXIS
ax.xaxis.set_major_locator(dates.MonthLocator(bymonth=None, bymonthday=1, interval=1, tz=None))
ax.xaxis.set_major_formatter(dates.DateFormatter('%a-%B-%d'));
```

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-12-31'],ylim=[50,65],grid=True,figsize=(12,6))
ax.set(xlabel='')
# SET THE TICK LOCATOR AND FORMATTER FOR THE MAJOR AXIS
ax.xaxis.set_major_locator(dates.MonthLocator(bymonth=None, bymonthday=1, interval=1, tz=None))
ax.xaxis.set_major_formatter(dates.DateFormatter('%d-%b-%y'));
```

#### MINOR AXIS

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-12-31'],ylim=[50,65],figsize=(12,6))
ax.set(xlabel='')
# SET THE TICK LOCATOR AND FORMATTER FOR THE MAJOR AXIS
ax.xaxis.set_major_locator(dates.MonthLocator(bymonth=None, bymonthday=1, interval=1, tz=None))
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%d-%b-%y'))

ax.xaxis.set_minor_locator(dates.DayLocator(bymonthday=15, interval=1, tz=None))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

ax.xaxis.grid(True)
ax.yaxis.grid(True);
```
