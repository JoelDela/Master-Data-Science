# Error Trend Seasonality Decomposition
- Error
- Trend
- Seasonality

Visualizing the data based on its decomposition is a good way to build an understanding of its behaviour

### Additive Models:
- Tendencia es mas lineal
- Tendencia y Seasonality constantes en el tiempo

### Multiplicative Models:
- Cuando incrementamos o decrementamos de manera no lineal

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
airline = pd.read_csv('../UPDATE_TSA_NOTEBOOKS/Data/airline_passengers.csv',index_col='Month',parse_dates=True)
airline.head()
```

```python
airline = airline.dropna()
```

```python
airline.plot(figsize=(16,6));
```

### Statsmodel Error Trend Seasonality Decomposition

```python
# Importando Libreria para descomposicin de la series temporal
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'],model='multiplicative')
```

```python
type(result)
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 16,6
result.plot();
```
