# Stationary Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../UPDATE_TSA_NOTEBOOKS/Data/samples.csv', index_col=0,parse_dates=True)
df.head()
```

### How Does a Stationary Series look like? 

- Constant Average
- Constant Variance
- Autocovariance does not depend on the time

_No muestra tendencia ni estacionalidad_

```python
df['a'].plot(figsize=(12,6));
```

### How does a Non Stationary Series look like?

- Muestra Tendencia
- Muestra Estacionalidad

```python
df['b'].plot(figsize=(12,6));
```

### Differencing Data

- To make a Time Series Stationary

```python
df['b'].plot(figsize=(12,6),title='Non Stationary Data');
```

#### First Difference Version

```python
from statsmodels.tsa.statespace.tools import diff
```

```python
# df['b'] - df['b'].shift(1)
```

```python
diff(df['b'],k_diff=1).plot(figsize=(12,6),title='Time Series Differenced');
```
