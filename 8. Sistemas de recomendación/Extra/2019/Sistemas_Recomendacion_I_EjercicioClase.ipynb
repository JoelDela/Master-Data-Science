```python
# voy a poner un texto y quiero que me retorne una lista de las películas que contienen este texto generando una 
# función.

def returnItemId(text, ids):
    #ids es el diccionario donde tiene que buscar
    
    text_= text.lower() #para evitar discrepancias con mayúsculas
    print(text_)
    
    
    search= [(k, v.lower().find(text_))
            for k,v in list(ids.items())]
    
    index= [k for k, v in search if v>-1]
    out=[]
    
    for i in index:
        out.append((i, ids[i]))
        
    return out # out es la lista de las películas que contienen el texto.
```

```python
# Detectamos que hay dos películas repetidas
# quiero un diccionario que me diga el nombre de la película y 
# al lado la lista de id en las que aparece, en caso de tener más de uno será una película que esté repetida
```

```python
duplicates_item_dict= {}

for k, v in list(item_dict.items()):
    if v in duplicates_item_dict:
        duplicates_item_dict[v].append(k)
    else:
        duplicates_item_dict[v]=[k]
```

```python
#veamos cuáles tienen más de de 1 id
```

```python
for i in duplicates_item_dict:
    if len(duplicates_item_dict[i])> 1:
        print(i, duplicates_item_dict[i])
```

```python
# Las películas con más de 1 id, hay que asignarles sólo 1. Hemos conseguido un nuevo índice
# que contiene los nombres y el número de línea en el que aparecían.
```

```python
for i, j in enumerate(duplicates_item_dict):
    print(i,j,duplicates_item_dict[j])
```

```python
unique_id_item_dict={}
for i, j in enumerate(duplicates_item_dict):
    for k in duplicates_item_dict[j]:
        unique_id_item_dict[k]=i
```
