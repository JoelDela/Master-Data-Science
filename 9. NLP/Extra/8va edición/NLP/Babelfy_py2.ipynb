To use pybabelfy you must register at http://babelfy.org/ and obtain your RESTful key

```
from pybabelfy.babelfy import *

text= "BabelNet es tanto un diccionario enciclopédico multilingüe como una red semántica"
lang = "SP"
key = "racatA44"

babelapi = Babelfy()

semantic_annotations = babelapi.disambiguate(text,lang,key)
```

```
for x in semantic_annotations:
  print x.dbpedia_url()
  print x.babelnet_url()
  print x.postag()
  print x.babelfy_dict.
```
