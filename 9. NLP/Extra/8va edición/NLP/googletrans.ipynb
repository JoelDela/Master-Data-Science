```
from googletrans import Translator
```

```
from googletrans import LANGCODES
```

```
for keys,values in LANGCODES.items():
    print(keys)
    print(values)
```

```
translator = Translator(service_urls=[
      'translate.google.com',
      'translate.google.co.kr',
    ])
```

```
trad = translator.translate('안녕하세요.')
```

```
print(trad.origin, ' -> ', trad.text)
print('language: ', trad.src, '/ pronunciation: ', trad.pronunciation)
```

```
trad = translator.translate('안녕하세요.', dest='ja')
```

```
print(trad.origin, ' -> ', trad.text)
print('language: ', trad.src, '/ pronunciation: ', trad.pronunciation)
```

```
translations = translator.translate(['The quick brown fox', 'jumps over', 'the lazy dog'], dest='es')
```

```
for translation in translations:
  print(translation.origin, ' -> ', translation.text)
```
