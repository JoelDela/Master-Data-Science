```
import textacy
```

```
import spacy
```

```
import cld2
```

```
text = 'Arthur Schopenhauer nació el 22 de febrero de 1788 en el seno de una acomodada familia de Danzig. El padre de Arthur, Heinrich Floris Schopenhauer, fue un próspero comerciante que inició a su hijo en el mundo de los negocios, haciéndole emprender largos viajes por Francia e Inglaterra. Su madre, Johanna Henriette Trosenier, fue una escritora que alcanzó cierta notoriedad al organizar soirées (veladas) literarias en la ciudad de Weimar'
```

```
isReliable, textBytesFound, details = cld2.detect(text)
```

```
print('  reliable: %s' % (isReliable != 0))
print('  textBytes: %s' % textBytesFound)
print('  details: %s' % str(details))
```

```
textacy.preprocess_text(text, lowercase=True, no_punct=True)
```

```
textacy.text_utils.KWIC(text, 'en', window_width=35)
```

```
doc = textacy.Doc(text, lang=u'es')
```

```
doc
```

```
es = textacy.load_spacy('es', disable=('parser',))
```

```
 list(textacy.extract.ngrams(doc, 3, filter_stops=True, filter_punct=True, filter_nums=False))
```

```
list(textacy.extract.ngrams(doc, 2, min_freq=2))
```

```
list(textacy.extract.named_entities(doc, drop_determiners=True))
```

```
pattern = textacy.constants.POS_REGEX_PATTERNS['en']['NP']
```

```
pattern
```

```
list(textacy.extract.pos_regex_matches(doc, pattern))
```

```
import textacy.keyterms
```

```
textacy.keyterms.textrank(doc, normalize='lemma', n_keyterms=10)
```
