Podemos crear un archivo nuevo:
    - 'a': significa ''append'' y sirve para añadir los que pongas después
    - 'w': significa ''write'' y sirve para reescribir lo que pongas después
    - 'r': que viene de ''read'' y es para leer. Es la que usa por defecto

```python
j=open('prueba.txt','a')
```

```python
j.write('Esto es para hacer una prueba')
```

```python
j.close()
f.close()
! cat prueba.txt
```

```python
k=open('prueba.txt','w')
```

```python
k.write('Y esta es una segunda prueba.')
```

```python
k.close()
! cat prueba.txt
```

```python
j=open('prueba.txt','a')
j.write('Añadimos la tercera prueba')
```

```python
j.close()
! cat prueba.txt
```
