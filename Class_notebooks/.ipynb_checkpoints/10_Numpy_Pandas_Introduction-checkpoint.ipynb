El comando np.linspace crea un vector desde -X hasta Y, cada Z intervalos

En plt.plot puedes decidir que tipo de linea quieres dibuja:
- Por ejemplo 'b.-' dibuja una línea en color 'blue' con líneas y puntos '.-'

```python
X= np.linspace(-np.pi, +np.pi, 25)
C=np.cos(X)
S=np.sin(X)
plt.plot(X,S,color='red')
plt.plot(X,C,'b.-') #Color azul con líneas y puntos
plt.grid(True) #Dibuja una rejilla. Es útil solo para encontrar puntos y demás
```

Una scatter plot es un dibujo de puntos. Sirve para datos de variables aleatorias correlacionadas.

```python
plt.scatter(X,C,color='red')
plt.scatter(X,S,color='blue')
```

__EJERCICIO__: Create a sample of points that follow the equation Y=AX+B where A=2.5 and B=20. Then, plot it as a cloud of points and a line

```python
X=np.linspace(-1.5,1.5,26)
A=2.5
B=20

x=np.random.uniform(-1.5,1.5,25)

plt.subplot(2,2,1)
plt.xlim(-1.5,1.5)
plt.ylim(16,25)
plt.plot(X,A*X+B,'r')
plt.subplot(2,2,2)
plt.xlim(-1.5,1.5)
plt.ylim(16,25)
plt.scatter(X,A*X+B,color='blue')
plt.subplot(2,2,3)
plt.xlim(-1.5,1.5)
plt.ylim(16,25)
plt.plot(x,A*x+B,'g')
plt.subplot(2,2,4)
plt.xlim(-1.5,1.5)
plt.ylim(16,25)
plt.scatter(x,A*x+B,color='yellow')
```

Los operadores bitwise son:
- x << y : Returns x with the bits shifted to the left by y places (and new bits on the right-hand-side are zeros). This is the same as multiplying x by 2**y.
- x >> y : Returns x with the bits shifted to the right by y places. This is the same as //'ing x by 2**y.
- x & y : Does a "bitwise and". Each bit of the output is 1 if the corresponding bit of x AND of y is 1, otherwise it's 0.
- x | y : Does a "bitwise or". Each bit of the output is 0 if the corresponding bit of x AND of y is 0, otherwise it's 1.
- ~ x : Returns the complement of x - the number you get by switching each 1 for a 0 and each 0 for a 1. This is the same as -x - 1.
- x ^ y : Does a "bitwise exclusive or". Each bit of the output is the same as the corresponding bit in x if that bit in y is 0, and it's the complement of the bit in x if that bit in y is 1.

```python
evens= a2%2==0
gt_3=a2 > 2
evens_and_gt3 = evens & gt_3
```

__EJERCICIO__: In a chicken and rabbit there are 35 heads and 94 legs. How many chickens and how many rabbits do we have?

Each chicken will have 1 head and 2 legs. Each rabbit will have 1 head and 4 legs

The equations will be:

$ head*chicken+head*rabbit=35$

$ legs*chicken+legs*rabbit=94$

Se puede pasar a matriz y resolverlo:



```python
A=np.mat([(1,1),(4,2)])
B=np.mat([(35),(94)]).reshape(2,1)

X=A.I*B
a,b=X
print(int(a),int(b))
```

```python
la.solve(A,B)
```

```python
A=np.array([(1,1),(4,2)])
B=np.array([(35),(94)])

la.solve(A,B)
```

__Es muy importante conocer la principal diferencia entre arrays y series. En un array, la suma de dos valores se produce SEGÚN SU POSICIÓN, es decir, dos arrays que esten descolocados no sumará "correctamente" (en caso de que necesitasemos que sumase de forma concreta). En cambio en Series, como se encuentra definido por el índice, lo suma según este índice, por lo que no es necesario ni que esté en orden ni que se cumplan las dimensiones, a diferencia que el array__

### Dropping entries from an axis

```python
s5=pd.Series(np.arange(5),list('jduvk'))
s5
```

```python
s4=s5.drop(['d','k'])
print(s4)
print(s5)
```

Vemos que la particularidad de .drop es que no se modifica el original. Para modificar el original habría que usar el parámetro "inplace=True"

```python
s5.drop('d',inplace=True)
s5
```
