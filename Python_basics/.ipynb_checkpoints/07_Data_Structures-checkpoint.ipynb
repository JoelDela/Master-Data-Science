## Bucles y funciones

```python
def add_one(value):
    return value+1
def add_two(value):
    return value+2
def add_three(value):
    return value+3
```

```python
math_ops=[add_one,add_two,add_three]
```

```python
def math_values(values,ops):
    results=[]
    for value in values:
        for function in ops:
            new_value=function(value)
            results.append(new_value)
    return results    
        
```

```python
math_values([1,2,3],math_ops)
```

### Función MAP

Se encarga de aplicar una función a cualquier secuencia

```python
l=[1,2,3]
```

```python
for i in map(add_one,l):
    print(i)
```

__Ejercicio: Quedarnos solo con los números pares de la lista a y elevarlos al cuadrado__

```python
new_list=[]
for i in a:
    if i%2==0 and i!=0:
        new_list.append(i**2)
    else:
        pass
print(new_list)
```

```python
[i**2 for i in a if i%2==0 and i!=0]
```

```python
a3=[value**2 if value%2==0 else 1 for value in a]
a3
```

Calcular el cuadrado de los valores de a usando map y lambda

```python
a4=list(map(lambda x: x**2,a))
a4
```

__EJERCICIO__: Conseguir una lista en mayúsculas de todas las palabras de más de dos caracteres

### Nested list comprehension

```python
list_of_lists=[[1,2,3],[4,5,6],[7,8,9]]
```

```python
one_list=[]
for par in list_of_lists:
    for n in par:
        one_list.append(n)
one_list    
```

El bucle hay que poner igual tanto en un bucle normal como en una list comprehension

```python
other_list=[n for par in list_of_lists if 2 in par for n in par]
```

```python
other_list
```

__EJERCICIO__: Lista con 10 nombres. Poner las vocales con mayúsculas y las consonantes con minúsculas

```python
names=['Antonio','Pedro','Marta','Laura','Ana','Daniel','Jose','Borja','Cristina','Javier']
```

```python
new_names=[]
for name in names:
    new_name=''
    for i in name:
        if i.lower() in 'aeiou':
            new_name+=i.upper()
        else:
            new_name+=i.lower()
    new_names.append(new_name)
new_names
```
